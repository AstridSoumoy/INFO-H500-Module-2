```python
from skimage.io import imread,imshow,imsave
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
%matplotlib inline


im = imread('walking.jpg')

## -- Your code here -- ##
lut = np.zeros((256,)) #create empty array with zero
for v in range(0,256):
    lut[v] = 255-v #fill the array from 255 to 0
new_im = lut[im]
plt.figure(figsize=(8,4))
plt.subplot(1,2,1) #supblot is useful to display images side-by-side
plt.imshow(im, cmap=cm.gray)
plt.title('Original')
plt.subplot(1,2,2)
plt.imshow(new_im, cmap=cm.gray)
plt.title('LUT(Original)')
plt.show()
```

```python
from skimage.io import imread,imshow,imsave
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import math
%matplotlib inline


im = imread('walking.jpg')

## -- Your code here -- ##
lut = np.zeros((256,)) #create empty array with zero
for v in range(0,256):
    lut[v] = math.floor((v*7)/255) #fill the array from 255 to 0
new_im = lut[im]
plt.figure(figsize=(8,4))
plt.subplot(1,2,1) #supblot is useful to display images side-by-side
plt.imshow(im, cmap=cm.gray)
plt.title('Original')
plt.subplot(1,2,2)
plt.imshow(new_im, cmap=cm.gray)
plt.title('LUT(Original)')
plt.show()
```

```python
from skimage.io import imread,imshow,imsave
import numpy as np
from PIL import ImageFilter, Image
import matplotlib.pyplot as plt
from matplotlib import cm
%matplotlib inline


im = Image.open('walking.jpg')

## -- Your code here -- ##
kernel = (-1, -1, -1, -1, 9, -1, -1, -1, -1)

new_im = im.filter(ImageFilter.Kernel((3,3), kernel, 1, 0))

plt.figure(figsize=(8,4))
plt.subplot(1,2,1) #supblot is useful to display images side-by-side
plt.imshow(im, cmap=cm.gray)
plt.title('Original')
plt.subplot(1,2,2)
plt.imshow(new_im, cmap=cm.gray)
plt.title('LUT(Original)')
plt.show()
```

```python
from skimage.filters.rank import mean, median
from skimage.io import imread,imshow,imsave
from skimage.morphology import disk
import matplotlib.pyplot as plt
%matplotlib inline

im = imread('astronaut_noisy.jpg')

## -- Your code here -- ##
selem = disk(2)
new_im = median(im, selem)

plt.figure(figsize=(10,10))
plt.subplot(1,2,1) #supblot is useful to display images side-by-side
plt.imshow(im, cmap=cm.gray)
plt.title('Original')
plt.subplot(1,2,2)
plt.imshow(new_im, cmap=cm.gray)
plt.title('LUT(Original)')
plt.show()
```
